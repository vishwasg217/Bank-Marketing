# Outlier Detection

```python
import pandas as pd
from sklearn.ensemble import IsolationForest

# Load data

# Define features and target variable
X = df.drop('target_encoded', axis=1)

# Instantiate Isolation Forest algorithm
isof = IsolationForest(n_estimators=100, contamination=0.1)

# Fit the algorithm to the data
isof.fit(X)

# Predict outliers
outliers = isof.predict(X)

# Print number of outliers detected
print("Number of outliers:", sum(outliers == -1))
outliers.shape
```

```python
df['is_inliner'] = outliers
df = df[df['is_inliner'] == 1]

```

```python
df.drop('is_inliner', axis=1, inplace=True)
df.shape
```

```python
X_train, X_test, y_train, y_test = dp.split_data(df, 'target_encoded')
```

```python
bag_clf = BaggingClassifier( DecisionTreeClassifier(), n_estimators=500, max_samples=100, bootstrap=True, n_jobs=-1) 
bag_clf.fit(X_train, y_train) 
y_pred = bag_clf.predict(X_test)
bag_clf.score(X_test, y_test)
```

```python
rnd_clf = RandomForestClassifier(n_estimators=500, n_jobs=-1) 
rnd_clf.fit(X_train, y_train) 
y_pred_rf = rnd_clf.predict(X_test)
rnd_clf.score(X_test, y_test)
```

```python
from sklearn.ensemble import AdaBoostClassifier 
ada_clf = AdaBoostClassifier( DecisionTreeClassifier(max_depth=1), n_estimators=500, algorithm="SAMME.R") 
ada_clf.fit(X_train, y_train)
y_pred_ada = ada_clf.predict(X_test)
ada_clf.score(X_test, y_test)
```

```python
from sklearn.ensemble import GradientBoostingRegressor 
gbrt = GradientBoostingRegressor(n_estimators=500) 
gbrt.fit(X_train, y_train)
y_pred_grad = gbrt.predict(X_test)
gbrt.score(X_test, y_test)
```

# feature scaling

```python
X = df.drop('target_encoded', axis=1)
y = df['target_encoded']
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

# PCA

```python
pca = PCA(n_components=0.85)
X_pca = pca.fit_transform(X_scaled)
X_pca.shape, X.shape
```

# grid search

```python
from sklearn.model_selection import GridSearchCV
clf = GridSearchCV(RandomForestClassifier(), {
    'n_estimators': [50, 250, 500],
    'max_depth': [4, 8, 16]
}, cv=5, return_train_score=False)
clf.fit(X_pca, y)
res = pd.DataFrame(clf.cv_results_)
```
